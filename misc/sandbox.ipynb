```python
import numpy as np
import matplotlib.pyplot as plt

def draw_matrix(matrix, cell_size=1.0, font_size=10):
    n = len(matrix)
    fig, ax = plt.subplots(figsize=(n * cell_size, n * cell_size), dpi=500)
    ax.set_xlim(0, n)
    ax.set_ylim(0, n)

    ax.set_xticks([])
    ax.set_yticks([])
    
    # Рисуем сетку и заполняем значения
    for i in range(n):
        for j in range(n):
            # Квадратик
            rect = plt.Rectangle((j, n - 1 - i), 1, 1, fill=True, edgecolor='black', facecolor='white', lw=2)
            ax.add_patch(rect)
            # Текст (значение матрицы)
            val = matrix[i, j]
            text = f"{val:.3f}" if isinstance(val, (float, np.floating)) else str(val)
            ax.text(j + 0.5, n - 0.5 - i, text, ha='center', va='center', fontsize=font_size)
    
    plt.gca().invert_yaxis()  # Чтобы матрица выглядела как в математической нотации
    plt.box(False)
    plt.show()
```

```python
import math

def distance_matrix(rows, cols):
    # Создаем пустую матрицу
    matrix = [[0 for _ in range(cols)] for _ in range(rows)]
    
    # Координаты центра верхней правой ячейки
    center_x = cols - 0.5
    center_y = 0.5
    
    for i in range(rows):
        for j in range(cols):
            # Координаты центра текущей ячейки
            cell_x = j + 0.5
            cell_y = i + 0.5
            
            # Вычисляем расстояние между центрами
            distance = math.sqrt((cell_x - center_x)**2 + (cell_y - center_y)**2)
            
            # Округляем до двух знаков после запятой
            rounded_distance = round(distance, 2)
            
            # Для целых чисел убираем .0
            if rounded_distance == int(rounded_distance):
                matrix[i][j] = int(rounded_distance)
            else:
                matrix[i][j] = rounded_distance
    
    return matrix

# Пример использования
rows, cols = 2, 2
result = distance_matrix(rows, cols)
for row in result:
    print(row)
```

```python
import math

def distance_matrix(rows, cols):
    # Создаем пустую матрицу
    matrix = [[0 for _ in range(cols)] for _ in range(rows)]
    
    # Координаты центра верхней ЛЕВОЙ ячейки (теперь (0.5, 0.5))
    center_x = 0.5
    center_y = 0.5
    
    for i in range(rows):
        for j in range(cols):
            # Координаты центра текущей ячейки
            cell_x = j + 0.5
            cell_y = i + 0.5
            
            # Вычисляем расстояние между центрами
            distance = math.sqrt((cell_x - center_x)**2 + (cell_y - center_y)**2)
            
            # Округляем до двух знаков после запятой
            rounded_distance = round(distance, 5)
            
            # Для целых чисел убираем .0
            if rounded_distance == int(rounded_distance):
                matrix[i][j] = int(rounded_distance)
            else:
                matrix[i][j] = rounded_distance
    
    return matrix

# Пример использования
rows, cols = 50, 50
result = distance_matrix(rows, cols)
for row in result:
    print(row)
```

```python
def percentage_diff_matrix(matrix1, matrix2):
    """
    Вычисляет матрицу процентных отличий между двумя матрицами.
    
    Параметры:
    matrix1 -- первая матрица (список списков)
    matrix2 -- вторая матрица (список списков)
    
    Возвращает:
    Матрицу, где каждый элемент - это процентное отличие между соответствующими элементами
    входных матриц, вычисленное как: |v1 - v2| / ((v1 + v2)/2) * 100%
    """
    # Проверяем, что матрицы одинаковой формы
    if len(matrix1) != len(matrix2) or any(len(row1) != len(row2) for row1, row2 in zip(matrix1, matrix2)):
        raise ValueError("Матрицы должны иметь одинаковую форму")
    
    result = []
    for row1, row2 in zip(matrix1, matrix2):
        result_row = []
        for v1, v2 in zip(row1, row2):
            # Вычисляем процентное отличие
            if v1 == 0 and v2 == 0:
                # Оба нуля - отличие 0%
                diff = 0.0
            elif v2 == 0:
                diff = float('inf')
            else:
                diff = v1 / v2
            result_row.append(round(diff, 3))
        result.append(result_row)
    
    return result

# Пример использования

# print(result)
# print(travel_time)
diff_matrix = percentage_diff_matrix(travel_time, result)
for row in diff_matrix:
    print(row)
```

```python
def relative_diff_matrix(matrix1, matrix2):
    """
    Вычисляет матрицу относительных отличий между двумя матрицами.
    Возвращает значения в долях (0.01 = 1%).
    """
    # Проверяем совпадение размеров матриц
    if len(matrix1) != len(matrix2):
        raise ValueError("Разное количество строк в матрицах")
    
    result = []
    for i in range(len(matrix1)):
        if len(matrix1[i]) != len(matrix2[i]):
            raise ValueError(f"Разная длина строки {i} в матрицах")
        
        row = []
        for j in range(len(matrix1[i])):
            a = matrix1[i][j]
            b = matrix2[i][j]
            
            # Вычисляем относительное отличие
            if a == 0 and b == 0:
                row.append(0.0)
            else:
                avg = (abs(a) + abs(b)) / 2
                if avg == 0:
                    row.append(float('inf'))
                else:
                    diff = abs(a - b) / avg
                    row.append(round(diff, 4))  # Округляем до 4 знаков
        result.append(row)
    
    return result



diff_matrix = relative_diff_matrix(travel_time, result)

diff_matrix
```

```python
def generate_cells(start_point, k, n, m, cell_size=100):
    R = 6371000  # радиус Земли в метрах
    deg_per_meter_lat = 1 / 111320  # примерно 1 градус широты на 111.32 км

    result = []

    for dz in range(k):
        for dy in range(n):
            for dx in range(m):
                # Смещение в метрах
                d_depth = start_point.depth + dz * cell_size
                d_lat_m = dy * cell_size
                d_lon_m = dx * cell_size

                # Широта смещается на юг (уменьшается)
                new_lat = start_point.lat - d_lat_m * deg_per_meter_lat

                # Долгота зависит от текущей широты
                deg_per_meter_lon = 1 / (111320 * math.cos(math.radians(new_lat)))
                new_lon = start_point.lon + d_lon_m * deg_per_meter_lon

                result.append([d_depth, new_lat, new_lon])

    return result
```

```python
import math

# WGS84 параметры
a = 6378137.0          # Большая полуось (экваториальный радиус)
f = 1 / 298.257223563  # Сплюснутость
b = a * (1 - f)        # Малая полуось
e_sq = f * (2 - f)     # Эксцентриситет в квадрате

def geodetic_to_ecef(lat, lon, h):
    lat_rad = math.radians(lat)
    lon_rad = math.radians(lon)
    N = a / math.sqrt(1 - e_sq * math.sin(lat_rad) ** 2)

    x = (N + h) * math.cos(lat_rad) * math.cos(lon_rad)
    y = (N + h) * math.cos(lat_rad) * math.sin(lon_rad)
    z = (N * (1 - e_sq) + h) * math.sin(lat_rad)

    return x, y, z

def ecef_to_geodetic(x, y, z):
    # Метод Bowring
    eps = e_sq / (1.0 - e_sq)
    p = math.sqrt(x**2 + y**2)
    theta = math.atan2(z * a, p * b)
    lon = math.atan2(y, x)
    lat = math.atan2(z + eps * b * math.sin(theta)**3,
                     p - e_sq * a * math.cos(theta)**3)
    N = a / math.sqrt(1 - e_sq * math.sin(lat)**2)
    h = p / math.cos(lat) - N

    return math.degrees(lat), math.degrees(lon), h

def enu_to_ecef(dx, dy, dz, ref_lat, ref_lon, ref_h):
    lat_rad = math.radians(ref_lat)
    lon_rad = math.radians(ref_lon)

    t = [
        [-math.sin(lon_rad), -math.sin(lat_rad) * math.cos(lon_rad), math.cos(lat_rad) * math.cos(lon_rad)],
        [math.cos(lon_rad), -math.sin(lat_rad) * math.sin(lon_rad), math.cos(lat_rad) * math.sin(lon_rad)],
        [0, math.cos(lat_rad), math.sin(lat_rad)]
    ]

    x_offset = t[0][0]*dx + t[0][1]*dy + t[0][2]*dz
    y_offset = t[1][0]*dx + t[1][1]*dy + t[1][2]*dz
    z_offset = t[2][0]*dx + t[2][1]*dy + t[2][2]*dz

    x0, y0, z0 = geodetic_to_ecef(ref_lat, ref_lon, ref_h)
    return x0 + x_offset, y0 + y_offset, z0 + z_offset

def generate_cells_precise(start_point, k, n, m, cell_size=30000):
    """
    Точная генерация ячеек с учётом кривизны Земли.
    Возвращает список координат [depth, lat, lon]
    """
    result = []
    for dz in range(k):
        for dy in range(n):
            for dx in range(m):
                # В ENU: X = восток, Y = север, Z = вверх
                local_east = dx * cell_size
                local_north = -dy * cell_size  # юг — отрицательное направление
                local_up = -dz * cell_size     # вниз = отрицательное "вверх"

                x, y, z = enu_to_ecef(local_east, local_north, local_up,
                                      start_point.lat, start_point.lon, -start_point.depth)
 
                lat, lon, alt = ecef_to_geodetic(x, y, z)
                depth = -alt
                result.append([depth, lat, lon])
    return result
```

```python
class Point:
    def __init__(self, lat, lon, depth):
        self.lat = lat
        self.lon = lon
        self.depth = depth

start = Point(lat=32.5, lon=120.1, depth=0)
cells = generate_cells_precise(start, k=10, n=10, m=10)

for c in cells:
    print(c)
```
